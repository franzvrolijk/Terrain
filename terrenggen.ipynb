```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import generate
import random
from matplotlib.pyplot import figure
%matplotlib inline


dim = 20 #Dimensjoner/proposjoner for data

xdim = dim
ydim = dim
passes = 5 #Antall smoothing-passes
random.seed(350) #Endrer seed for å få forskjellig resultat hver generering

fig = plt.figure() #Ny figur
ax1 = fig.add_subplot(111, projection='3d') #Definerer et plt-plot

A = 100 * generate.genHeightMap(xdim, ydim, passes) #Fyller A med høydedata, * 100 for å få verdier mellom 1 og 100
A = np.log(A)**3 #Tar logaritmen for å gjøre forskjeller mindre ekstreme, **3 for å gi mer forskjell i høyde
generate.remEdges(A) #Fjerner kanter, da genHeightMap gir støyete/u-smoothede verdier på kantene


# Genererer akser for plotting, etc:
x = np.array([[i] * xdim for i in range(ydim)]).ravel() # x coordinates of each bar
y = np.array([i for i in range(xdim)] * ydim) # y coordinates of each bar
z = np.zeros(ydim*xdim) # z coordinates of each bar
dx = np.ones(ydim*xdim) # length along x-axis of each bar
dy = np.ones(ydim*xdim) # length along y-axis of each bar
dz = A.ravel() # length along z-axis of each bar (height)

ax1.bar3d(x, y, z, dx, dy, dz) #Genererer 3d-bar plot

fig.set_size_inches(20, 10, forward=True) #Endrer størrelse på figuri tommer, kan endre ppi

fig.savefig('test.png') #Saver til høyoppløst bilde
```
