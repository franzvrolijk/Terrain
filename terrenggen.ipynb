```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import generate
import random
from matplotlib.pyplot import figure
%matplotlib inline


dim = 30

xdim = dim
ydim = dim
passes = 4
random.seed(350)

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

A = 100 * generate.genHeightMap(xdim, ydim, passes)
A = np.log(A)**3
generate.remEdges(A)

x = np.array([[i] * xdim for i in range(ydim)]).ravel() # x coordinates of each bar
y = np.array([i for i in range(xdim)] * ydim) # y coordinates of each bar
z = np.zeros(ydim*xdim) # z coordinates of each bar
dx = np.ones(ydim*xdim) # length along x-axis of each bar
dy = np.ones(ydim*xdim) # length along y-axis of each bar
dz = A.ravel() # length along z-axis of each bar (height)

ax1.bar3d(x, y, z, dx, dy, dz)


fig.set_size_inches(40, 20, forward=True)



fig.savefig('test.png')
```
